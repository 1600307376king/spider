```python
from sklearn.linear_model import SGDRegressor, LinearRegression, Ridge
from sklearn.preprocessing import PolynomialFeatures

x = np.arange(0, 1000, 1).reshape(-1, 1)
y = np.array(df.iloc[:1000, 1:2].values.tolist())

poly_reg =PolynomialFeatures(degree=1)
X_ploy =poly_reg.fit_transform(x)
lin_reg=LinearRegression()
lin_reg.fit(X_ploy,y)
y_pred = lin_reg.predict(X_ploy)


df1 = pd.DataFrame(y_pred, index=np.arange(0, 1000, 1), columns=['y'])

df2 = pd.DataFrame(y, index=np.arange(0, 1000, 1), columns=['y_p'])
df3 = pd.concat([df1, df2], axis=1)
df3.iplot(kind='lines')
```
