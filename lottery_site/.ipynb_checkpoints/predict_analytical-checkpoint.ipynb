```python
import random


def get_number(max_number, epoch):
    pellet = list(range(1, max_number + 1))
    random.shuffle(pellet)
    choice_list = []
    for i in range(epoch):
        a = random.choice(pellet)
        pellet.remove(a)
        choice_list.append(a)
    choice_list.sort()
    return choice_list

red = get_number(33, 6)
blue = get_number(16, 1)
print(red, blue)

history_list = []
for k in range(1000):
    history_list.append(get_number(33, 6) + (get_number(16, 1)))
    
# dfx = pd.DataFrame(history_list)
# dfx.iloc[30:60, 0:8].iplot(kind='lines')
```
